### D05 感測器資料自動蒐集
### 作業1 : 練習使用一個民生物聯網 API，例如空氣、地震等感測站所回傳的紀錄資料。
參考資料網址： https://ci.taiwan.gov.tw/dsp/environmental.aspx

```python
import requests
import pandas as pd

# 國家空品站，最新一筆 PM2.5 感測資料
url = 'https://sta.ci.taiwan.gov.tw/STA_AirQuality_v2/v1.0/Datastreams?$expand=Thing,Observations($orderby=phenomenonTime%20desc;$top=1)&$filter=name%20eq%20%27PM2.5%27%20and%20Thing/properties/authority%20eq%20%27%E8%A1%8C%E6%94%BF%E9%99%A2%E7%92%B0%E5%A2%83%E4%BF%9D%E8%AD%B7%E7%BD%B2%27%20and%20substringof(%27%E7%A9%BA%E6%B0%A3%E5%93%81%E8%B3%AA%E6%B8%AC%E7%AB%99%27,Thing/name)&$count=true'
res = requests.get(url)
jd1 = res.json()

d1_arr, d2_arr, d3_arr, d4_arr, d5_arr = [], [], [], [], []
for i in range(len(jd1['value'])) :
    d1 = jd1['value'][i]['Thing']['properties']['authority']
    d1_arr.append(d1)
    d2 = jd1['value'][i]['Thing']['properties']['stationName']
    d2_arr.append(d2)
    d3 = jd1['value'][i]['Thing']['properties']['stationID']
    d3_arr.append(d3)
    d4 = jd1['value'][i]['Observations'][0]['result'] # PM2.5 value
    d4_arr.append(d4)
    d5 = jd1['value'][i]['Observations'][0]['phenomenonTime']
    d5_arr.append(d5)
    df1 = pd.DataFrame({'authority':d1_arr, 'stationName':d2_arr, 'stationID':d3_arr, 'PM2.5':d4_arr, 'phenomenonTime':d5_arr})
df1
```

### 作業2 : 練習操作 OGC SensorThings API，將環保局測站位置的資料抓取下來，並且觀察下載資料的內容。
參考網址：https://sta.ci.taiwan.gov.tw/STA_AirQuality_EPAIoT/v1.0/Things

```python
# 環保局測站資料
url = 'https://sta.ci.taiwan.gov.tw/STA_AirQuality_EPAIoT/v1.0/Things'
res = requests.get(url)
jd2 = res.json()

import numpy as np
d1_arr, d2_arr, d3_arr, d4_arr, d5_arr, d6_arr = [], [], [], [], [], []
dic = {}
for i in range(len(jd2['value'])) :
    d1 = jd2['value'][i]['properties']['authority']
    d1_arr.append(d1)
    if 'city' not in jd2['value'][i]['properties'].keys() :
        d2 = np.nan
        d2_arr.append(d2)
    else :
        d2 = jd2['value'][i]['properties']['city']
        d2_arr.append(d2)
    if 'township' not in jd2['value'][i]['properties'].keys() :
        d3 = np.nan
        d3_arr.append(d3)
    else :
        d3 = jd2['value'][i]['properties']['township']
        d3_arr.append(d3) 
    d4 = jd2['value'][i]['properties']['stationID']
    d4_arr.append(d4)
    d5 = jd2['value'][i]['properties']['locationId']
    d5_arr.append(d5)
    d6 = jd2['value'][i]['properties']['Description']
    d6_arr.append(d6)
    df2 = pd.DataFrame({'authority':d1_arr, 'city':d2_arr, 'township':d3_arr, 'stationID':d4_arr, 'locationId':d5_arr, 
                        'Description':d6_arr})
df2
```
